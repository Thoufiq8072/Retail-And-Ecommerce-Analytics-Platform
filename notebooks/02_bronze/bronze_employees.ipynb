### Importing Needed Modules

```python
import sys
import os
from pyspark.sql.functions import *
from pyspark.sql.types import *

## Replace the workspace path with your own
sys.path.append(f"/Workspace/Users/mohammedthoufiq9360@gmail.com/Retail-And-Ecommerce-Analytics-Platform")
from src.paths import RAW_EMPLOYEES_PATH, BRONZE_EMPLOYEES_PATH
from src.utils import add_ingestion_timestamp
```

### Reading the Raw Employees File

```python
emp_df = spark.read.format("csv")\
    .option("header", "true")\
    .option("inferSchema", "true")\
    .load(RAW_EMPLOYEES_PATH)
```

```python
display(emp_df.limit(5))
```

### Adding metadata columns ("ingestion_ts", "_source_file")

```python
emp_df = add_ingestion_timestamp(emp_df)
```

### Creating Bronze Employees Table

```python
(
    emp_df
    .write
    .format("delta")
    .option("delta.columnMapping.mode", "name")
    .mode("overwrite")
    .saveAsTable(BRONZE_EMPLOYEES_PATH)
)
```
