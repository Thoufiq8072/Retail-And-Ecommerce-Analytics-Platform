### Importing the needed Modules

```python
import sys
import os
from pyspark.sql.functions import *
from pyspark.sql.types import *

## Replace the workspace path with your own
sys.path.append(f"/Workspace/Users/mohammedthoufiq9360@gmail.com/Retail-And-Ecommerce-Analytics-Platform")
from src.paths import RAW_CUSTOMERS_PATH, BRONZE_CUSTOMERS_PATH
from src.utils import add_ingestion_timestamp
```

### Reading the Raw Customers Files

```python
cust_df = spark.read.format("csv")\
    .option("header", "true")\
    .option("inferSchema", "true")\
    .load(RAW_CUSTOMERS_PATH)
```

```python
display(cust_df.limit(5))
```

### Adding metadata Columns ("ingestion_ts", "_source_file")

```python
cust_df = add_ingestion_timestamp(cust_df)
```

### Creating Bronze **Customers** Table

```python
(
    cust_df
    .write
    .format("delta")
    .option("delta.columnMapping.mode", "name")
    .mode("overwrite")
    .saveAsTable(BRONZE_CUSTOMERS_PATH)
)
```
