```python
import sys
import os
from pyspark.sql.functions import *
from pyspark.sql.types import *

## Replace the workspace path with your own
sys.path.append(f"/Workspace/Users/mohammedthoufiq9360@gmail.com/Retail-And-Ecommerce-Analytics-Platform")
from src.paths import RAW_DISCOUNTS_PATH, BRONZE_DISCOUNTS_PATH
from src.utils import add_ingestion_timestamp
```

```python
disc_df = spark.read.format("csv")\
    .option("header", "true")\
    .option("inferSchema", "true")\
    .load(RAW_DISCOUNTS_PATH)
```

```python
display(disc_df.limit(5))
```

```python
disc_df = add_ingestion_timestamp(disc_df)
```

```python
(
    disc_df
    .write
    .format("delta")
    .option("delta.columnMapping.mode", "name")
    .mode("overwrite")
    .saveAsTable(BRONZE_DISCOUNTS_PATH)
)
```
