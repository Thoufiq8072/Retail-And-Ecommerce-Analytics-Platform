### Importing the Needed Modules

```python
import sys
import os
from pyspark.sql.functions import *
from pyspark.sql.types import *

## Replace the workspace path with your own
sys.path.append(f"/Workspace/Users/mohammedthoufiq9360@gmail.com/Retail-And-Ecommerce-Analytics-Platform")
from src.paths import RAW_STORES_PATH, BRONZE_STORES_PATH
from src.utils import add_ingestion_timestamp
```

### Reading the Raw Stores files

```python
store_df = spark.read.format("csv")\
    .option("header", "true")\
    .option("inferSchema", "true")\
    .load(RAW_STORES_PATH)
```

```python
display(store_df.limit(5))
```

### Adding metadata columns ("ingestion_ts", "_source_file")

```python
store_df = add_ingestion_timestamp(store_df)
```

### Creating Bronze Stores Table

```python
(
    store_df
    .write
    .format("delta")
    .option("delta.columnMapping.mode", "name")
    .mode("overwrite")
    .saveAsTable(BRONZE_STORES_PATH)
)
```

```python
spark.read.table(BRONZE_STORES_PATH).limit(5).display()
```
